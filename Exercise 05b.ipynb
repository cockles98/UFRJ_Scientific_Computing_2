Apresentação 5, encontre a região de estabilidade do método RK4. Pode ser necessário escrever um código para desenhar a região.

```python
import numpy as np
import matplotlib.pyplot as plt

X, Y = np.meshgrid(np.arange(-3, 3.01, 0.01), np.arange(-3, 3.01, 0.01))
Mu = X + 1j * Y
F = abs(1 + Mu + (Mu**2) / 2 + (Mu**3) / 6 + (Mu**4) / 24)

# Create a contour chart with the inner area filled
plt.contour(X, Y, F, levels=[1], colors='k')
plt.contourf(X, Y, F, levels=[0, 1], colors='blue', alpha=0.5)

plt.xlabel('Re(Lambda*h)')
plt.ylabel('Im(Lambda*h)')
plt.title('Stability Region of RK4')
plt.grid()
```
