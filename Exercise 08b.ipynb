```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha, beta, T_inf = 15e-4,5.6e-5,50

Nx = 100
hx = 1/(Nx+1)

Tf = 10
ht = hx*hx/(8*alpha)
Nt = (int)(Tf/ht)

T = np.zeros((Nx+2,Nt+1))

sx = alpha*ht/hx/hx
```

```python
for n in range(Nt):
    #BND COND
    T[0,n] = 0
    T[Nx+1,n] = -2*hx*beta*(T[Nx,n]**4 - T_inf**4) + T[Nx-1,n]

    for i in range(0,Nx+1):
        T[i,n+1] = T[i,n] + sx*(T[i+1,n] - 2*T[i,n] + T[i-1,n])

plt.plot(T[:,1223])
```
