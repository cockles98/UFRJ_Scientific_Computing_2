#**3.3.13**

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of complex values
X, Y = np.meshgrid(np.arange(-3, 3.01, 0.01), np.arange(-3, 3.01, 0.01))
Mu = X + 1j * Y

# Calculate the R function
R = 1 + Mu + 0.5 * Mu**2

# Calculate the absolute value of R
Rhat = np.abs(R)

# Create a contour plot
plt.contour(X, Y, Rhat, levels=[1], colors='k')

# Create a filled contour plot for the area inside the contour
plt.contourf(X, Y, Rhat, levels=[0, 1], colors='brown', alpha=0.5)

plt.xlabel('Re(Lambda*h)')
plt.ylabel('Im(Lambda*h)')
plt.title('Stability Region of RK2')
plt.grid(True)
plt.show()
```

#**3.3.13**

```python
import numpy as np
import matplotlib.pyplot as plt

def CN(initial_condition:float = 1, h:float = 0.1, initial_time:float = 0, max_time:float = 10, lambda_value:float = -5):
       t_n = [initial_time]   # time values
       u_n = [initial_condition]   # numerical values
       u_tn = [initial_condition]   # analitycal values
       e_n = [0]   # error values
       while t_n[-1] < max_time:
           t_n.append(t_n[-1]+h)
           u_n.append((u_n[-1]*(1-(abs(lambda_value)*h)/2))/(1+(abs(lambda_value)*h)/2))
           u_tn.append(np.exp(t_n[-1]*lambda_value))
           e_n.append(abs(u_n[-1]-u_tn[-1]))
       return t_n, u_n, u_tn, e_n, h

t, u1, u2, e, h = CN()

plt.plot(t,u1,label='numerical')
plt.plot(t,u2,label='analytical')
plt.plot(t,e,label='error')
plt.grid()
plt.legend()
```
