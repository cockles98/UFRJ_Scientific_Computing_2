```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(u1, u2):
    u_1 = 1.2 * u1 - 0.6 * u1 * u2
    u_2 = 0.8 * u2 + 0.3 * u1 * u2
    return u_1, u_2
```

```python
u1 = [2]
u2 = [1]
h = 1e-3
T = [0]

while T[-1] < 20:
  k11, k12 = f(u1[-1], u2[-1])
  k21, k22 = f(u1[-1]+((h/2)*k11), u2[-1]+((h/2)*k12))
  k31, k32 = f(u1[-1]+((h/2)*k21), u2[-1]+((h/2)*k22))
  k41, k42 = f(u1[-1]+(h*k31), u2[-1]+(h*k32))

  u1_new = u1[-1] + (h/6)*(k11+2*k21+2*k31+k41)
  u2_new = u2[-1] + (h/6)*(k12+2*k22+2*k32+k42)

  u1.append(u1_new)
  u2.append(u2_new)
  T.append(T[-1]+h)
```

```python
plt.plot(T,u1, label='u1')
plt.xlabel('Time')
plt.ylabel('u')
plt.title('U_1 Graph')
plt.legend()
```

```python
plt.plot(T,u2, label='u2', color='r')
plt.xlabel('Time')
plt.ylabel('u')
plt.title('U_2 Graph')
plt.legend()
```
