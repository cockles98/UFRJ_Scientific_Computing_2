```python
import matplotlib.pyplot as plt
```

```python
def FE(t_0:float = 0, u_0:float = 1, v_0:float = 0, h:float = 0.1, max_t:float = 1000, eps:float = 0.1):
    '''Forward Euler, returns the t_n, u_n, v_n lists and the values h and eps.
    You can set the initial and final times (t_0 and max_t), the initial
    conditions (u_0 and v_0), the size of each step (h) and the eps'''
    t_n = [t_0]
    u_n= [u_0]
    v_n= [v_0]
    f1_n = [v_0]
    f2_n = [(-eps)*((u_0**2)-1)*v_0-u_0]
    while t_n[-1] <= max_t:
        t_n.append(t_n[-1]+h)   # updating time
        u_n.append(u_n[-1]+(v_n[-1]*h))
        v_n.append(v_n[-1]+h*((-eps)*((u_n[-1]**2)-1)*v_n[-1]-u_n[-1]))
        f1_n.append(v_n[-1])
        f2_n.append(-eps*((u_n[-1]**2)-1)*v_n[-1]-u_n[-1])
    return t_n, u_n, v_n, h, eps
```

```python
'Plotting the graph'
for i in [0.01,0.1,0.5,1,1.5,2,2.5,3,3.5,4]:
    t_n, u_n, v_n, h, eps = FE(eps=i)
    plt.plot(u_n[-70:-1],v_n[-70:-1])   # plotting only the convergent curve

plt.legend()
plt.grid()
plt.title('fixing h=0.1 and varying the eps=[0.01,0.1,0.5,1,1.5,2,2.5,3,3.5,4]')
plt.xlabel('u_n')
plt.ylabel('v_n')
```
