Equação de Burger, todos os métodos (com volume finito)

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML
```

```python
N = 100
hx = 1/(N+1)
a = 1
ht = (0.1/abs(a))*hx
sigma = a*(ht/hx)

Tf = 0.4
Ntempo = int(Tf/ht)

x = np.linspace(0,1,N+2)
u = np.zeros((Ntempo+1,N+2))

for i in range(0,N+2):
    if x[i]<0.5:
        u[0,i] = 0
    else:
        u[0,i] = 1
```

```python
for n in range(Ntempo):
    u[n,0],u[n,N+1] = 0,1
    for i in range(1,N+1):
        u[n+1,i] = (1-sigma)*u[n,i] + sigma*u[n,i-1]
```

```python
for n in range(Ntempo):
    u[n,0],u[n,N+1] = 0,1
    for i in range(1,N+1):
        u[n+1,i] = u[n,i] - (sigma/2)*(u[n,i+1]-u[n,i-1])
```

```python
for n in range(Ntempo):
    u[n,0],u[n,N+1] = 0,1
    for i in range(1,N+1):
        u[n+1,i] = (1+sigma)*u[n,i] - sigma*u[n,i+1]
```

```python
for n in range(Ntempo):
    u[n,0],u[n,N+1] = 0,1
    for i in range(1,N+1):
        ap,an = max(a,0),min(a,0)
        uxp,uxn = (u[n,i]-u[n,i-1])/hx,(u[n,i+1]-u[n,i])/hx
        u[n+1,i] = u[n,i] - ht*(ap*uxp+an*uxn)
```

```python
# Lax-Friedrich method
for n in range(Ntempo):
    u[n,0],u[n,N+1] = 0,1
    for i in range(1,N+1):
        u[n+1,i] = (u[n,i-1]+u[n,i+1])/2 - (sigma/2)*(u[n,i+1]-u[n,i-1])
```

```python
# Leapfrog method
for n in range(Ntempo):
    u[n,0],u[n,N+1] = 0,1
    for i in range(1,N+1):
        if n==0:
            ap,an = max(a,0),min(a,0)
            uxp,uxn = (u[n,i]-u[n,i-1])/hx,(u[n,i+1]-u[n,i])/hx
            u[n+1,i] = u[n,i] - ht*(ap*uxp+an*uxn)
        else:
            u[n+1,i] = u[n-1,i] - sigma*(u[n,i+1]-u[n,i-1])
```

```python
# Lax-Wendroff method
for n in range(Ntempo):
    u[n,0],u[n,N+1] = 0,1
    for i in range(1,N+1):
        u[n+1,i] = u[n,i]-(sigma/2)*(u[n,i+1]-u[n,i-1])+((sigma**2)/2)*(u[n,i+1]-2*u[n,i]+u[n,i-1])
```

```python
fig, ax = plt.subplots()
ax.set_xlim(( 0, 1))
ax.set_ylim((-0.5, 1.5))
line, = ax.plot([], [], lw=2)

def init():
    line.set_data([], [])
    return (line,)

def animate(i):
    x = np.linspace(0, 1, N+2)
    y = u[i,:]
    line.set_data(x, y)
    return (line,)

anim = animation.FuncAnimation(fig, animate, init_func=init,frames=Ntempo, interval=1, blit=True)
plt.close(fig)
HTML(anim.to_html5_video())
```

```python
N = 400
hx = 1.0/N
ht = 0.5*hx
Tf = 180*ht
Nt = int(Tf/ht)

u = np.zeros((Nt+1,N+1))

for i in range(N+1):
    xi = i*hx
    u[0,i] = np.cos(2*np.pi*xi) + 1
```

```python
for n in range(Nt):
    u[n,0] = u[n,N-1]
    u[n,N] = u[n,1]
    for i in range(1,N):
        u[n+1,i] = u[n,i] - u[n,i]*(ht/hx)*(u[n,i]-u[n,i-1])
```

```python
fig, ax = plt.subplots()
ax.set_xlim(( 0, 10))
ax.set_ylim((-0.1, 2.1))
line, = ax.plot([], [], lw=2)

def init():
    line.set_data([], [])
    return (line,)

def animate(i):
    x = np.linspace(0, 10, N+1)
    y = u[i,:]
    line.set_data(x, y)
    return (line,)

anim = animation.FuncAnimation(fig, animate, init_func=init,frames=Nt, interval=20, blit=True)
plt.close(fig)
HTML(anim.to_html5_video())
```
