```python
import numpy as np
import matplotlib.pyplot as plt
```

# Example 1

```python
def func(x):
    return np.sin(np.pi*x)
```

```python
N = 20
h = 1.0/(N+1)

u_left = 0
u_right = 1

x = np.zeros(N)
u_numerical = np.zeros(N)
u_exact = np.zeros(N)
f = np.zeros(N)
A = np.zeros((N,N))

for i in range(1,N+1):
    x[i-1] = i*h
    f[i-1] = h* h* func(x[i-1])
    u_exact[i-1] = (func(x[i-1])/np.pi/np.pi) + x[i-1]
    if i>1:
        A[i-1,i-2] = -1
    else:
        f[i-1] = f[i-1] + u_left
    A[i-1,i-1] = 2.0
    if i<N:
        A[i-1,i] = -1
    else:
        f[i-1] = f[i-1] + u_right

u_numerical = np.linalg.solve(A,f)
```

```python
plt.plot(x, u_numerical, label="Numerical")
plt.plot(x, u_exact, label="Exact")
plt.grid()
plt.legend();
```

```python
N = 100
h = 1.0/(N+1)

u_left = 0
u_right = 1

x = np.zeros(N)
u_numerical = np.zeros(N)
u_exact = np.zeros(N)
f = np.zeros(N)
A = np.zeros((N,N))
eps = 0.01
a = (-eps/h + 1/2)
b = (2*eps/h)
c = (-eps/h - 1/2)

for i in range(1,N+1):
    x[i-1] = i*h
    f[i-1] = h* h* func(x[i-1])
    u_exact[i-1] = (np.exp(x[i-1]/eps) - 1)/ (np.exp(1/eps) -1)
    if i>1:
        A[i-1,i-2] = c
    else:
        f[i-1] = f[i-1] -c* u_left
    A[i-1,i-1] = b
    if i<N:
        A[i-1,i] = a
    else:
        f[i-1] = f[i-1] -a*u_right

u_numerical = np.linalg.solve(A,f)
```
